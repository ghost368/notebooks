```python
from qplot import get_axes, set_pd_backend, set_mpl_layout, set_plotly_layout
import qplot.activate

import plotly.express as px
import pandas as pd
import numpy as np
import scipy as sp
```

```python
rng = np.random.default_rng()
data = pd.DataFrame(rng.standard_normal(size=(1000, 5)), columns=list('abcde'))
```

```python
ser = data['a']
vals = ser.to_numpy()
```

```python
fig = px.histogram(vals, nbins=100)
set_plotly_layout(fig)
fig.show()
```
