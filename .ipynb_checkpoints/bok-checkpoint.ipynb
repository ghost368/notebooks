```python
from matplotlib import pyplot as plt
import numpy as np
import math 
#needed for definition of pi
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.tools as tls

# redefine some plotting methods for Nextjournal support

plt.show = plt.gcf
```

```python
import pandas as pd

df = pd.DataFrame(np.random.rand(10, 10), index=pd.date_range('2010-01-01', '2011-01-01')[:10], columns=list('abcdefghjk'))
```

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Note that using plt.subplots below is equivalent to using
# fig = plt.figure and then ax = fig.add_subplot(111)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13,5))
ax1.plot(t, s)

ax1.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax1.grid()

ax1.plot(t, t)
df.plot(ax=ax2, legend=False)
ax2.grid()

tls.mpl_to_plotly(fig)
```
